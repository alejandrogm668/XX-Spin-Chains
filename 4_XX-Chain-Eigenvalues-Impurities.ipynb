```python
import numpy as np
from scipy.sparse.linalg import eigsh
import matplotlib.pyplot as plt
from Functions import *
import time
```

```python
n_sites = 10
J = 1.0
h = 1.0 
δh = 5.0
distance = 0

H = get_sparse_impure_hamiltonian(n_sites,J,h,δh,distance)
H
```

```python
# Lowest eigenvalue 

eigvals = eigsh(H,k=1,which='SA',return_eigenvectors=False)
eigvals
```

```python
# Distribution of eigenvalues for different distances

plt.figure(figsize=(8,8))

for i in range(3):
    H = get_sparse_impure_hamiltonian(n_sites,J,h,δh,distance=2*i)
    eigvals = eigsh(H,k=2**(n_sites)-1,which='SA',return_eigenvectors=False)
    plt.plot(eigvals,'.',label="distance = "+str(2*i))
    
plt.legend()    
```

```python
# Distribution of eigenvalues for different δh

plt.figure(figsize=(8,8))

fields = [5.0,10.0,20.0]

for δh in fields:
    H = get_sparse_impure_hamiltonian(n_sites,J,h,δh,distance=2)
    eigvals = eigsh(H,k=2**(n_sites)-1,which='SA',return_eigenvectors=False)
    plt.plot(eigvals,'.',label="δh = "+str(δh))
    
plt.legend()    
```
